```python
import cv2
import numpy as np
import yaml
import matplotlib.pyplot as plt
import math

yaml_file = "arena3.yaml"

with open(yaml_file, "r") as f:
    meta = yaml.safe_load(f)

resolution = meta["resolution"]  
pgm_file = meta["image"]

gray = cv2.imread(pgm_file, cv2.IMREAD_GRAYSCALE)
if gray is None:
    raise RuntimeError(f"Failed to load {pgm_file}")

binary = cv2.inRange(gray, 0, 0)   # PURE black → white
img = cv2.imread(pgm_file, 0)

# --- Edge Detection ---
edges = cv2.Canny(img, 50, 150, apertureSize=3)

# --- HoughLinesP ---
lines = cv2.HoughLinesP(
    edges,
    rho=10,
    theta=np.pi / 180,
    threshold=80,
    minLineLength=5,
    maxLineGap=1
)


blurred_image = cv2.GaussianBlur(img, (6, 6), 0) # (kernel_size, kernel_size), sigmaX
# --- Draw Lines ---
output = cv2.cvtColor(blurred_image, cv2.COLOR_GRAY2BGR)

if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(output, (x1, y1), (x2, y2), (0, 0, 255), 2)
else:
    print("No lines detected!")

# --- Show Results ---
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Edges")
plt.imshow(blurred_image, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("HoughLinesP")
plt.imshow(output)
plt.axis("off")

plt.show()



```
